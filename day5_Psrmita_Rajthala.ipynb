```python
# Positive, Negative, or Zero (if-elif-else)
num = 5
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#Even or Odd (if-else)
num = 3
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#Voting Eligibility (if-else)
age = 18
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
#Grade Calculator (if-elif-else)
marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
    print("Grade C")
elif marks >= 60:
    print("Grade D")
else:
    print("Grade F")
```
