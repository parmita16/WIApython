```python
#hybrid inheritence 1
class A:
    def __init__(self, pname):
        self.pname = pname
    def display(self):
        return f"Hii I am parent {self.pname}"
class B(A):
    def __init__(self, pname, cname):
        A.__init__(self, pname)
        self.cname = cname
    def display(self):
        print(A.display(self))
        return f"Hii I am child of A = {self.cname}"
class C(A):
    def __init__(self, pname, gcname):
        A.__init__(self, pname)
        self.gcname = gcname
    def display(self):
        print(A.display(self))
        return f"Hii I am another child of A = {self.gcname}"
class D(B, C):
    def __init__(self, pname, cname, gcname,dname):
        B.__init__(self, pname, cname)
        C.__init__(self, pname, gcname)
        self.dname = dname
    def display(self):
        print(B.display(self))
        print(C.display(self))
        return f"Hii I am grandchild {self.dname}"
print(D.mro())
```

```python
d = D("Parmita", "Sneha", "Bandana","Feint")
print(d.display())
```

```python
#hybrid inheritence 2
class A:
    def __init__(self, pname):
        self.pname = pname
    def display(self):
        return f"Hii I am parent {self.pname}"
class B:
    def __init__(self, cname):
        self.cname = cname
    def display(self):
        return f"Hii I am 2nd parent {self.cname}"
class C(A, B):
    def __init__(self, pname, cname, gcname):
        A.__init__(self, pname)
        B.__init__(self, cname)
        self.gcname = gcname
    def display(self):
        print(A.display(self))
        print(B.display(self))
        return f"Hii I am child of A and B = {self.gcname}"
class D(C):
    def __init__(self, pname, cname, gcname, dname):
        C.__init__(self, pname, cname, gcname)
        self.dname = dname
    def display(self):
        print(C.display(self))
        return f"Hii I am grandchild {self.dname}"
print(D.mro())
```

```python
d = D("Parmita", "Sneha", "Bandana","Feint")
print(d.display())
```

```python
#hybrid inheritence 3
class A:
    def __init__(self, pname):
        self.pname = pname
    def display(self):
        return f"Hii I am parent {self.pname}"
class B:
    def __init__(self, cname):
        self.cname = cname
    def display(self):
        return f"Hii I am 2nd parent {self.cname}"
class C(A):
    def __init__(self, pname, gcname):
        A.__init__(self, pname)
        self.gcname = gcname
    def display(self):
        print(A.display(self))
        return f"Hii I am child of A = {self.gcname}"
class D(B):
    def __init__(self, cname, gcname):
        B.__init__(self, cname)
        self.gcname = gcname
    def display(self):
        print(B.display(self))
        return f"Hii I am child of B = {self.gcname}"
class E(C, D):
    def __init__(self, pname, cname, gcname):
        C.__init__(self, pname, gcname)
        D.__init__(self, cname, gcname)
    def display(self):
        print(C.display(self))
        print(D.display(self))
        return f"Hii I am child of C and D = {self.gcname}"
print(E.mro())
```

```python
#1. Inheritance and Method Overriding
class Shape:
    def __init__(self, name):
        self.name = name
    def area(self):
        return 0
    def info(self):
        return self.name
class Rectangle(Shape):
    def __init__(self, name, length, width):
        super().__init__(name)
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def info(self):
        print("Name:", self.name)
        print("Length:", self.length)
        print("Width:", self.width)
        print("Area:", self.area())
r = Rectangle("Rectangle", 10, 5)
r.info()
```

```python
#2. Employee : SoftwareEngineer & DevopsDeveloper
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    def calc_bonus(self):
        return self.salary * 0.10
    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)
class SoftwareEngineer(Employee):
    def __init__(self, name, salary, designation):
        super().__init__(name, salary)
        self.designation = designation
    def calc_bonus(self):
        bonus = self.salary * 0.20
        self.display()
        print("Designation:", self.designation)
        print("Bonus:", bonus)
class DevopsDeveloper(Employee):
    def __init__(self, name, salary, designation):
        super().__init__(name, salary)
        self.designation = designation
    def calc_bonus(self):
        bonus = self.salary * 0.15
        self.display()
        print("Designation:", self.designation)
        print("Bonus:", bonus)
s = SoftwareEngineer("blue", 50000, "Software Engineer")
s.calc_bonus()
print()
d = DevopsDeveloper("Ocean", 50000, "DevOps Developer")
d.calc_bonus()
```

```python
#3. Multiple Inheritance
class Flyable:
    def fly(self):
        print("Can Fly")
class Swimmable:
    def swim(self):
        print("Can Swim")
class AmphibiousVehicle(Flyable, Swimmable):
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed
    def fly(self):
        print(self.name, "is flying at", self.speed, "km/h")
    def swim(self):
        print(self.name, "is swimming at", self.speed, "km/h")
    def perform_actions(self):
        self.fly()
        self.swim()
v = AmphibiousVehicle("SeaPlane", 200)
v.perform_actions()
```
