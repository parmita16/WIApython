```python
#Reverse a list 
a = [10, 20, 30, 40, 50]
a = a[::-1]
print(a)
```

```python
#swap first & last
a[0], a[-1] = a[-1], a[0]
print(a)
```

```python
#symmetric difference
a = [1, 2, 3, 5, 6, 7, 8, 9]
b = [1, 3, 5, 7, 9]
print(a)
print(b)
```

```python
s1 = set(a)
s2 = set(b)
print(s1)
print(s2)
```

```python
c = list(s1 ^ s2)
print(c)
```

```python
print(list (s1.symmetric_difference(s2)))
```

```python
#moving number of index 2 to last
a = [5, 10, 15, 20]
b = a.pop(2)
print(b)
```

```python
#Merge two dictionaries
store1 = {"apple": 5, "banana": 3}
store2 = {"orange": 4, "banana": 10}
store1.update(store2)
print(store1)
```
