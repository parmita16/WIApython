```python
# capitalize() - Makes the first letter uppercase.
text = "hello world"
print(text.capitalize())
```

```python
# casefold() - Converts all letters to lowercase (stronger than lower()).
text = "HELLO"
print(text.casefold())
```

```python
# center() - Centers the string within the given width.
text = "Hello"
print(text.center(15))
```

```python
# count() - Counts how many times a substring appears.
text = "apple apple"
print(text.count("apple"))
```

```python
# encode() - Converts the string into bytes.
text = "Hello"
print(text.encode())
```

```python
# endswith() - Checks if the string ends with the given text.
text = "hello.py"
print(text.endswith(".py"))
```

```python
# expandtabs() - Replaces tab spaces with normal spaces.
text = "Hello\tWorld"
print(text.expandtabs(4))
```

```python
# find() - Returns the first position of a substring.
text = "Hello World"
print(text.find("World"))
```

```python
# format() - Inserts values into placeholders.
name = "Alice"
print("Hello {}".format(name))
```

```python
# format_map() - Inserts values using a dictionary.
data = {"name": "Alice"}
print("Hello {name}".format_map(data))
```

```python
# index() - Returns the position of a substring.
text = "Hello World"
print(text.index("World"))
```

```python
# isalnum() - Checks if all characters are letters or numbers.
text = "Hello123"
print(text.isalnum())
```

```python
# isalpha() - Checks if all characters are letters.
text = "Hello"
print(text.isalpha())
```

```python
# isascii() - Checks if all characters are ASCII.
text = "Hello"
print(text.isascii())
```

```python
# isdecimal() - Checks if all characters are decimal numbers.
text = "123"
print(text.isdecimal())
```

```python
# isdigit() - Checks if all characters are digits.
text = "123"
print(text.isdigit())
```

```python
# isidentifier() - Checks if the string is a valid variable name.
text = "my_name"
print(text.isidentifier())
```

```python
# islower() - Checks if all letters are lowercase.
text = "hello"
print(text.islower())
```

```python
# isnumeric() - Checks if all characters are numeric.
text = "123"
print(text.isnumeric())
```

```python
# isprintable() - Checks if all characters can be printed.
text = "Hello"
print(text.isprintable())
```

```python
# isspace() - Checks if the string contains only spaces.
text = "   "
print(text.isspace())
```

```python
# istitle() - Checks if every word starts with a capital letter.
text = "Hello World"
print(text.istitle())
```

```python
# isupper() - Checks if all letters are uppercase.
text = "HELLO"
print(text.isupper())
```

```python
# join() - Joins list items into one string.
words = ["I", "Love", "Python"]
print(" ".join(words))
```

```python
# ljust() - Aligns text to the left.
text = "Hi"
print(text.ljust(10))
```

```python
# lower() - Converts all letters to lowercase.
text = "HELLO"
print(text.lower())
```

```python
# lstrip() - Removes spaces from the left side.
text = "   Hello"
print(text.lstrip())
```

```python
# maketrans() - Creates a translation table.
table = str.maketrans("a", "x")
print("apple".translate(table))
```

```python
# partition() - Splits the string into three parts.
text = "apple-banana"
print(text.partition("-"))
```

```python
# removeprefix() - Removes the given prefix.
text = "Mr.John"
print(text.removeprefix("Mr."))
```

```python
# removesuffix() - Removes the given suffix.
text = "photo.jpg"
print(text.removesuffix(".jpg"))
```

```python
# replace() - Replaces one text with another.
text = "I like cats"
print(text.replace("cats", "dogs"))
```

```python
# rfind() - Finds the last position of a substring.
text = "apple apple"
print(text.rfind("apple"))
```

```python
# rindex() - Returns the last position of a substring.
text = "apple apple"
print(text.rindex("apple"))
```

```python
# rjust() - Aligns text to the right.
text = "Hi"
print(text.rjust(10))
```

```python
# rpartition() - Splits the string from the last occurrence.
text = "apple-banana-orange"
print(text.rpartition("-"))
```

```python
# rsplit() - Splits the string from the right side.
text = "apple,banana,orange"
print(text.rsplit(","))
```

```python
# rstrip() - Removes spaces from the right side.
text = "Hello   "
print(text.rstrip())
```

```python
# split() - Splits the string into a list.
text = "apple banana orange"
print(text.split())
```

```python
# splitlines() - Splits the string into lines.
text = "Hello\nWorld"
print(text.splitlines())
```

```python
# startswith() - Checks if the string starts with the given text.
text = "hello.py"
print(text.startswith("hello"))
```

```python
# strip() - Removes spaces from both sides.
text = "   Hello   "
print(text.strip())
```

```python
# swapcase() - Swaps uppercase and lowercase letters.
text = "Hello World"
print(text.swapcase())
```

```python
# title() - Makes the first letter of every word uppercase.
text = "hello world"
print(text.title())
```

```python
# translate() - Replaces characters using a translation table.
table = str.maketrans("ae", "12")
text = "apple"
print(text.translate(table))
```

```python
# upper() - Converts all letters to uppercase.
text = "hello"
print(text.upper())
```

```python
# zfill() - Adds zeros to the beginning of the string.
text = "42"
print(text.zfill(5))
```

```python
# replace() - Replaces only the specified number of occurrences.
text = "apple apple apple"
print(text.replace("apple", "banana", 2))
```

```python
# split() - Splits only the specified number of times.
text = "apple banana orange mango"
print(text.split(" ", 2))
```

```python
# rsplit() - Splits from the right a specified number of times.
text = "apple banana orange mango"
print(text.rsplit(" ", 2))
```

```python
# center() - Centers the string using a custom fill character.
text = "Hi"
print(text.center(10, "*"))
```

```python
# ljust() - Left-aligns the string using a fill character.
text = "Hi"
print(text.ljust(10, "-"))
```

```python
# rjust() - Right-aligns the string using a fill character.
text = "Hi"
print(text.rjust(10, "-"))
```

```python
# strip() - Removes specific characters from both ends.
text = "###Hello###"
print(text.strip("#"))
```

```python
# lstrip() - Removes specific characters from the left side.
text = "###Hello"
print(text.lstrip("#"))
```

```python
# rstrip() - Removes specific characters from the right side.
text = "Hello###"
print(text.rstrip("#"))
```
