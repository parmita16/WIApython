```python
#Question 1: Rectangle Class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)
```

```python
rect1 = Rectangle(10, 5)
rect2 = Rectangle(8, 4)
rect3 = Rectangle(6, 3)
```

```python
print("Area of Rectangle 1 =", rect1.area())
print("Perimeter of Rectangle 2 =", rect2.perimeter())
```

```python
#Question 2: Person Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hi, I am {self.name}, and I am {self.age} years old.")
```

```python
person1 = Person("Parmita", 20)
person2 = Person("Bishesh", 23)
person1.introduce()
person2.introduce()
```

```python
#Question 3: Book Class
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    def details(self):
        print(f"Book: {self.title} by {self.author}")
```

```python
book1 = Book("lalalala", "bruh")
book2 = Book("blablabla", "bro")
book1.details()
book2.details()
```

```python
#Question 4: Circle Class
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius
    def circumference(self):
        return 2 * 3.14 * self.radius
```

```python
circle = Circle(5)
print("Area =", circle.area())
print("Circumference =", circle.circumference())
```
