```python
#Nested loop
students = [
    ["Parmita", 20, "A+"],
    ["B", 22, "B"],
    ["BM", 21, "A+"]
]

print(students)
```

```python
#accessing elements
print(students[0])      
print(students[1][0])   
print(students[2][2])
```

```python
students[1][2] = "A"   
print(students)
```

```python
# Delete 
del students[0][1]   
print(students)
```

```python
# Delete B
del students[1]   
print(students)
```

```python
students.pop(1)
print(students)
```

```python
#Nested dictionary
students = {
    "Parmita": {
        "age": 20,
        "grade": "A"
    },
    "B": {
        "age": 22,
        "grade": "B"
    }
}

print(students)
```

```python
#accessing elements
print(students["Parmita"])
print(students["B"]["grade"])
```

```python
students["B"]["grade"] = "A+"
print(students)
```

```python
del students["Parmita"]["grade"]
print(students)
```

```python
del students["B"]
print(students)
```

```python
students.pop("Parmita")
print(students)
```
