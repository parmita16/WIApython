```python
#Question 1: Calculate Area of a Rectangle
def calculate_area(length, width):
    return length * width

try:
    length = float(input("Enter length: "))
    width = float(input("Enter width: "))

    result = calculate_area(length, width)
    print("Area of Rectangle =", result)

except ValueError:
    print("Error: Please enter valid numeric values.")

finally:
    print("Program Finished.")
```

```python
#Question 2: Display Employee Information
def display_employee(**a):
    print("\nEmployee Information:")
    for key, value in a.items():
        print(key, ":", value)

try:
    display_employee(
        Name="Parmita",
        Age=28,
        Department="IT",
        Position="Graphic Designer",
        Salary=50000
    )

except Exception as e:
    print("Error:", e)

finally:
    print("Employee information displayed successfully.")
```

```python
#Question 3: Grade Calculator
def calculate_grade():
    subjects = ["Math", "Science", "English", "History", "Art"]
    total = 0
    try:
        for subject in subjects:
            marks = int(input(f"Enter marks for {subject}: "))
            if marks < 0 or marks > 100:
                raise ValueError("Marks must be between 0 and 100.")
            total += marks
        percentage = (total / 500) * 100
        if percentage >= 90:
            grade = "A"
        elif percentage >= 80:
            grade = "B"
        elif percentage >= 70:
            grade = "C"
        elif percentage >= 60:
            grade = "D"
        else:
            grade = "F"
        print("\nTotal Marks:", total)
        print("Percentage:", round(percentage, 2))
        print("Grade:", grade)
    except ValueError as e:
        print("Error:", e)
    finally:
        print("Grade calculation completed.")
calculate_grade()
```

```python
#Question 4: Even Numbers (1–100)
def even():
    try:
        print("Even Numbers from 1 to 100:")

        for i in range(1, 101):
            if i % 2 == 0:
                print(i)

    except Exception as e:
        print("Error:", e)

    finally:
        print("Program Finished.")

even()
```

```python
#Question 5: FizzBuzz
def fizzbuzz():
    try:
        for i in range(1, 101):

            if i % 3 == 0 and i % 5 == 0:
                print("FizzBuzz")
            elif i % 3 == 0:
                print("Fizz")
            elif i % 5 == 0:
                print("Buzz")
            else:
                print(i)

    except Exception as e:
        print("Error:", e)

    finally:
        print("FizzBuzz Program Completed.")

fizzbuzz()
```

```python
#Question 6: Multiplication Table (Odd Multiples Only)
def multiplication_table(number):
    try:
        for i in range(1, 11):

            if i % 2 == 0:
                continue

            print(number, "x", i, "=", number * i)

    except Exception as e:
        print("Error:", e)

    finally:
        print("Multiplication Table Completed.")

try:
    num = int(input("Enter a number: "))
    multiplication_table(num)

except ValueError:
    print("Error: Please enter a valid integer.")

finally:
    print("Program Finished.")
```

```python
#Question 7: Guessing Game
def guessing_game():
    secret_number = 25

    try:
        while True:
            guess = int(input("Guess the number: "))

            if guess < secret_number:
                print("Too Low!")
            elif guess > secret_number:
                print("Too High!")
            else:
                print("Congratulations! You guessed the correct number.")
                break

    except ValueError:
        print("Error: Please enter only numbers.")

    finally:
        print("Thanks for playing!")

guessing_game()
```
