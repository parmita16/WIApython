```python
#Q!
import asyncio
```

```python
async def sum_of_squares(name, number):
    result = 0
    for i in range(1, number + 1):
        print(f"Task {name}: Computing square of {i}...")
        result += i ** 2
        await asyncio.sleep(1)
    print(f"Task {name}: sum of squares({number}) = {result}")
    return result
```

```python
async def main():
    results = await asyncio.gather(
        sum_of_squares("A", 3),
        sum_of_squares("B", 5),
        sum_of_squares("C", 6)
    )
    print("Final Results:", results)
```

```python
await main()
```

```python
async def multiplication_table(name, number):
    for i in range(1, 6):
        print(f"Task {name}: Calculating {number} x {i}...")
        print(f"{number} x {i} = {number * i}")
        await asyncio.sleep(0.8)
    print(f"Task {name}: Completed multiplication table for {number}")
```

```python
async def main():
    await asyncio.gather(
        multiplication_table("A", 4),
        multiplication_table("B", 6),
        multiplication_table("C", 8)
    )
    print("All tables completed!")
```

```python
await main()
```

```python
#Q3
def execution_time(func):
    def wrapper(number):
        print("Function Started")
        result = func(number)
        print("Function Completed")
        return result
    return wrapper
```

```python
@execution_time
def calculate_square(number):
    return number * number
```

```python
result = calculate_square(7)
print("Returned Result:", result)
```
