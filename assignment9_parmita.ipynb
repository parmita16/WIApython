```python
#Q1
n = [1, 2, 3, 4, 5]
result = list(map(lambda x: x * 2, n))
print(result)
```

```python
#Q2
fruits = ["apple", "banana", "cherry"]
result = list(map(lambda x: x.upper(), fruits))
print(result)
```

```python
#Q3
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(2))
```

```python
#Q4
def string_len(s):
    if s == "":
        return 0
    return 1 + string_len(s[1:])

print(string_len("len"))
```
